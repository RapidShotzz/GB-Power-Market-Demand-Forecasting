```python
# converting settlement date to datetime format

df['settlement_date'] = pd.to_datetime(df['settlement_date'])
df.head()
```

```python
# creating a timestamp column from settlement date and settlement period

time_offset = pd.to_timedelta((df['settlement_period'] - 1) * 30, unit='m')
df['timestamp'] = df['settlement_date'] + time_offset
df.head()
```

```python
# setting the timestamp as index

df.set_index('timestamp', inplace=True)
df.head()
```

```python
# exploring impossible values (e.g. negative demand, wind generation greater than installed capacity)

df[df['england_wales_demand'] < 0]
```

```python
df[df['embedded_wind_generation'] > df['embedded_wind_capacity']]
```

```python
# extracting day, month, year, hour, minute, quarter and weekday and week number from timestamp

df.reset_index('timestamp', inplace=True)
df['day'] = df['timestamp'].dt.day
df['month'] = df['timestamp'].dt.month
df['year'] = df['timestamp'].dt.year
df['hour'] = df['timestamp'].dt.hour
df['minute'] = df['timestamp'].dt.minute
df['quarter'] = df['timestamp'].dt.quarter
df['weekday'] = df['timestamp'].dt.weekday
df['week_number'] = df['timestamp'].dt.isocalendar().week

df.head()
```

```python
# creating a column for name of the day (e.g. if weekday == 0 then Monday, if weekday == 1 then Tuesday etc)

def weekday_to_name(day):
    if day == 0:
        return 'Monday'
    elif day == 1:
        return 'Tuesday'
    elif day == 2:
        return 'Wednesday'
    elif day == 3:
        return 'Thursday'
    elif day == 4:
        return 'Friday'
    elif day == 5:
        return 'Saturday'
    else:
        return 'Sunday'

df['day_of_week'] = df['weekday'].apply(weekday_to_name)

df.head()
```

```python
# creating a column to flag weekends

def is_weekend(day):
    if day == 5 or day == 6:
        return True
    else:
        return False

df['is_weekend'] = df['weekday'].apply(is_weekend)

df.head()
```

```python
# creating a column for the morning peak hours (7am to 9am)

def is_morning_peak(hour):
    if hour >= 7 and hour <= 9:
        return True
    else:
        return False

df['is_morning_peak'] = df['hour'].apply(is_morning_peak)

df.head(30)
```

```python
# creating a column for the evening peak hours (16pm to 19pm)

def is_evening_peak(hour):
    if hour >= 16 and hour <= 19:
        return True
    else:
        return False

df['is_evening_peak'] = df['hour'].apply(is_evening_peak)

df.head(50)
```

```python
# creating a column for night hours (23pm to 6am)

def night_hours(hour):
    if hour <= 6 or hour == 23:
        return True
    else:
        return False

df['night'] = df['hour'].apply(night_hours)

df.head(50)
```

```python
# creating a column for lag demand (30 min)

df['lagged_demand_30m'] = df['england_wales_demand'].shift(1)
df.head()
```

```python
# creating a column for lag demand (1 hour)

df['lagged_demand_1h'] = df['england_wales_demand'].shift(2)
df.head()
```

```python
# creating a column for lag demand (24 hours)

df['lagged_demand_24h'] = df['england_wales_demand'].shift(48)
df.head(50)
```

```python
# creating a column for rolling mean demand (24 hours)

df['demand_rolling_mean_24h'] = df['england_wales_demand'].rolling(window=48, min_periods=1).mean()
df.head(20)
```

```python
# creating a column for renewable generation (wind + solar)

df['renewable_gen'] = df['embedded_wind_generation'] + df['embedded_solar_generation']
df.head(10)
```

```python
# creating a column for renewable percentage 

df['renewable_pct'] = df['renewable_gen'] / df['england_wales_demand']
df.head(10)
```

```python
# creating a column for net demand

df['net_demand'] = df['england_wales_demand'] - df['renewable_gen']
df.head(10)
```

```python
# can save the modified csv file 

df.head()
df.to_csv('cleaned_demand_data_2025.csv', index=False)
print('CSV filed saved successfully.')
```
